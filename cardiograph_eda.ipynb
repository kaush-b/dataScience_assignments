```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('Cardiotocographic.csv')
df.head()
```

```python
#checking for missing-values
missing_values=df.isnull().sum()
print(missing_values)
```

```python
rows_with_missing = df[df.isnull().any(axis=1)]
print(rows_with_missing)
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis', yticklabels=False)
plt.title('Heatmap of missing values')
plt.show()
```

```python
#replacing missing values with median values of the respective columns
df.fillna(df.median(), inplace=True)
df.isnull().sum()
```

```python
#correlation matrix
correlation_matrix = df.corr()
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, cmap='coolwarm', square=True, cbar_kws={"shrink": .8})
plt.title("correlation heatmap")
plt.show()
```

```python
#handling outliers using winsorization which replaces extreme values with the closest value within a certain percentile range
from scipy.stats.mstats import winsorize

# Define the percentage limits for winsorization (1% on each end)
limits = [0.01, 0.01]

# Apply winsorization to each column in the dataset
df_clean = df.copy()  # Create a copy of the dataframe
for col in df_clean.columns:  # Only apply to numerical columns
    df_clean[col] = winsorize(df_clean[col], limits=limits)

# Display the first few rows of the winsorized dataframe
df_clean.head()
```

```python
df_clean.describe()
```

```python
for col in df_clean.columns:
    plt.figure(figsize=(15,5))
    #histogram plot
    plt.subplot(1,2,1)
    sns.histplot(data=df_clean, x=col, kde=True)
    plt.title(f'Histogram of {col}')
```

```python
for col in df_clean.columns:
    plt.figure(figsize=(15,5))
    #violin plot
    plt.subplot(1,2,1)
    sns.violinplot(data=df_clean, x=col, kde=True)
    plt.title(f'violinplot of {col}')
```
